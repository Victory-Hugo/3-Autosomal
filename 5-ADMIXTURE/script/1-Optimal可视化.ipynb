```python
import matplotlib.pyplot as plt
from aquarel import load_theme
theme = load_theme('boxy_light')
theme.apply()
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
# Given data
Ks = list(range(10, 26))
cv_errors = [0.10008, 0.10023, 0.09946, 0.09946, 0.10123, 
             0.10073, 0.10007, 0.09990, 0.10154, 0.10190, 
             0.10142, 0.10310, 0.10319, 0.10141, 0.10250, 
             0.10140]

# Create the plot
plt.figure()
plt.plot(Ks, cv_errors, marker='o', linewidth=1)
plt.xlabel('K')
plt.ylabel('CV error')
plt.title('ADMIXTURE Cross validation')
plt.xticks(Ks)
plt.grid(False)
plt.tight_layout()
theme.apply_transforms()
plt.savefig('../output/K_selection.pdf')
plt.show()
```
