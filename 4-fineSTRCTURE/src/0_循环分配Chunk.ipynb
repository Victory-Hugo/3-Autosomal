```python
# 生成多个txt ID文件，要求每个ID文件必须包含DR列中为D的ID，并在DR列为R的ID随机抽取250个
# 最重要保证所有的DR列中为R的样本都被分配完
# 这一步是为了减轻计算压力，每次只计算一部分数据
# 将输出文件命名为Chunk_1，Chunk_2，Chunk_3，...，直到所有的ID都被分配完
# 保存到/mnt/d/幽门螺旋杆菌/Script/分析结果/6-fineSTRUCTURE/data/内
```

```python
import numpy as np
import os

# 获取D和R样本的ID
d_samples = df_META[df_META['DR'] == 'D']['ID'].tolist()
r_samples = df_META[df_META['DR'] == 'R']['ID'].tolist()

# 计算需要多少个chunk (每个chunk包含250个R样本)
chunk_size = 250
num_chunks = int(np.ceil(len(r_samples) / chunk_size))

# 创建输出目录
output_dir = '/mnt/d/幽门螺旋杆菌/Script/分析结果/6-fineSTRUCTURE/data/'
os.makedirs(output_dir, exist_ok=True)

# 随机打乱R样本的顺序
np.random.shuffle(r_samples)

# 生成每个chunk的文件
for i in range(num_chunks):
    chunk_num = i + 1
    
    # 获取这个chunk的R样本
    start_idx = i * chunk_size
    end_idx = min((i + 1) * chunk_size, len(r_samples))
    current_r_samples = r_samples[start_idx:end_idx]
    
    # 合并D样本和当前chunk的R样本
    all_samples = d_samples + current_r_samples
    
    # 将样本ID写入文件
    output_file = os.path.join(output_dir, f'Chunk_{chunk_num}.txt')
    with open(output_file, 'w') as f:
        for sample_id in all_samples:
            f.write(f'{sample_id}\n')
    
    print(f'Created Chunk_{chunk_num}.txt with {len(d_samples)} D samples and {len(current_r_samples)} R samples')
```
